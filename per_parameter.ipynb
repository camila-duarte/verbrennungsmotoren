* idleX.csv refers to the vehicle in the IDLE mode

* driveX.csv refers to the vehicle in motion mode in high speed roads

* liveX.csv refers to the vehicle in a specific traject (work --> home)

* ufpeX.csv refers to the vehicle in motion mode inside an university campus (low speed)

* longX.csv refers to long trips with the vehicle in motion mode.

```python
# importing of libaries
import numpy as np                      #functions for scientific computing, faster arrays
import pandas as pd                     #easy import and export of data, fast and easy-to-use datastructure
import math                             #mathematical functions
import matplotlib.pyplot as plt         #generation of Graphs
from scipy.constants import pi, g       #fundamental Constants
```

```python
# Path for every data set
path_data_drive1 = 'data/v2_drive1.csv'
path_data_idle1 = 'data/v2_idle1.csv'
path_data_live1 = 'data/v2_live1.csv'
path_data_long1 = 'data/v2_long1.csv'
path_data_ufpe1 = 'data/v2_ufpe1.csv'
path_data_long2 = 'data/long2.csv'

# Read all the data
data_drive1 = pd.read_csv(path_data_drive1)
data_idle1 = pd.read_csv(path_data_idle1)
data_live1 = pd.read_csv(path_data_live1)
data_long1 = pd.read_csv(path_data_long1)
data_long2 = pd.read_csv(path_data_long2)
data_ufpe1 = pd.read_csv(path_data_ufpe1)

# Imports the index

index_data_drive1 = np.arange(data_drive1.shape[0])
index_data_idle1 = np.arange(data_idle1.shape[0])
index_data_live1 = np.arange(data_live1.shape[0])
index_data_long1 = np.arange(data_long1.shape[0])
index_data_long2 = np.arange(data_long2.shape[0])
index_data_ufpe1 = np.arange(data_ufpe1.shape[0])

```

# Speed

```python
speed_data_drive1 = data_drive1['VEHICLE_SPEED ()']*3.6
speed_data_idle1 = data_idle1['VEHICLE_SPEED ()']*3.6
speed_data_live1 = data_live1['VEHICLE_SPEED ()']*3.6
speed_data_long1 = data_long1['VEHICLE_SPEED ()']*3.6
speed_data_ufpe1 = data_ufpe1['VEHICLE_SPEED ()']*3.6
```

```python
speed_drive1 = speed_data_drive1
speed_idle1 = speed_data_idle1
speed_live1 = speed_data_live1
speed_long1 = speed_data_long1
speed_ufpe1 = speed_data_ufpe1
```

```python
#plt.style.use('_mpl-gallery-nogrid')

plt.figure(figsize=(17, 5))
plt.plot(index_data_drive1, speed_drive1, label='drive',color="#4169e1")
plt.plot(index_data_idle1, speed_idle1, label='idle',color="#1e90ff")
plt.plot(index_data_live1, speed_live1, label='live',color="#000080")
plt.plot(index_data_long1, speed_long1, label='long',color="#003153")
plt.plot(index_data_ufpe1, speed_ufpe1, label='ufpe',color="#062a78")

plt.xlabel('Zeit in s')
plt.ylabel('Geschwindigkeit in kmh')
plt.legend()
plt.title('Geschwindigkeit der verschiedenen Fahrten')
plt.show()
```

# Throttle

```python
throttle_data_drive1 = data_drive1['THROTTLE ()']
throttle_data_idle1 = data_idle1['THROTTLE ()']
throttle_data_live1 = data_live1['THROTTLE ()']
throttle_data_long1 = data_long1['THROTTLE ()']
throttle_data_ufpe1 = data_ufpe1['THROTTLE ()']
```

```python
throttle_drive1 = throttle_data_drive1
throttle_idle1 = throttle_data_idle1
throttle_live1 = throttle_data_live1
throttle_long1 = throttle_data_long1
throttle_ufpe1 = throttle_data_ufpe1
```

```python
#plt.style.use('_mpl-gallery-nogrid')

plt.figure(figsize=(19, 5))
plt.plot(index_data_drive1,throttle_drive1, label='drive',color="#4169e1")
plt.plot(index_data_idle1, throttle_idle1, label='idle',color="#1e90ff")
plt.plot(index_data_live1, throttle_live1, label='live',color="#000080")
plt.plot(index_data_long1, throttle_long1, label='long',color="#003153")
plt.plot(index_data_ufpe1, throttle_ufpe1, label='ufpe',color="#062a78")

plt.xlabel('Zeit in s')
plt.ylabel('Drosselklappenstellung in Grad')
plt.title('Drosselklappenstellung der verschiedenen Fahrten')
plt.legend()
plt.show()
```
